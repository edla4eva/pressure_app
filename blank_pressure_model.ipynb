# Train Model

train the model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder

from sklearn.impute import SimpleImputer

# create an imputer object
imputer = SimpleImputer(strategy='mean')

# impute missing values in a column
df['P1'] = imputer.fit_transform(df[['P1']])
df['P2'] = imputer.fit_transform(df[['P2']])


# split data into training and test sets
train_df, test_df = train_test_split(df, test_size=0.2)

# prepare data for modeling
X_train = train_df.drop(columns=["P1"])
y_train = train_df["P1"]
X_test = test_df.drop(columns=["P1"])
y_test = test_df["P1"]

#encode datetime

# select the categorical feature to encode
# cat_feature = df["DateTime"]

# create a one-hot encoder and fit it to the data
# encoder = OneHotEncoder()
# cat_encoded = encoder.fit_transform(cat_feature.values.reshape(-1, 1))

# convert the sparse matrix to a dense matrix and add it to the DataFrame
# df_encoded = pd.concat([df, pd.DataFrame(cat_encoded.toarray(), columns=encoder.get_feature_names_out())], axis=1)

# drop the original categorical feature from the DataFrame
# df_encoded = df_encoded.drop(columns=["DateTime"])



# train a linear regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# evaluate model performance on test set
y_pred = lr_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean squared error:", mse)
```
